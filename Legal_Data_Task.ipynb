```python
# REMOVE STOPWORDS

nltk.download('stopwords')

from nltk.corpus import stopwords

removed_stopwords = []

stop_words = set(stopwords.words("english"))

for pdf_stopword in no_punct_pdfs:
  if isinstance(pdf_stopword, list):
    no_stopword = [token for token in pdf_stopword if token not in stop_words]
    removed_stopwords.append(no_stopword)
  else:
    print(f"unexpected data type: {type(pdf_stopword)}")
```

```python
# N-GRAMS

from nltk.util import ngrams

def generate_ngrams(removed_stopwords, n):
  n_grams = list(ngrams(removed_stopwords, n))
  return n_grams

n_grams_list = []

for ng in removed_stopwords:
  bi_grams = generate_ngrams(ng, 2)
  n_grams_list.append(bi_grams)
```

```python
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity

def calculate_similarity(embedding1, embedding2):

    embedding1 = embedding1 / np.linalg.norm(embedding1)
    embedding2 = embedding2 / np.linalg.norm(embedding2)

    similarity = np.dot(embedding1, embedding2)
    return similarity
```
